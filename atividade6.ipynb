Segunda questão

```python
import random
import numpy as np
from copy import deepcopy

k = 3
print(f'K escolhido igual a {k}.')
print()
axle_points_vector = []
class_points_vector = []
class_points_output = []
zeros = 0
ones = 0
for count in range(10):
  aux1 = random.uniform(1, 10)
  aux1 = round(aux1, 2)
  aux2 = random.uniform(1, 10)
  aux2 = round(aux2, 2)
  class_points_a = random.randint(0,1)
  class_points_b = random.randint(0,1)
  class_points_tuple = (class_points_a, class_points_b)
  class_points_vector.append(class_points_tuple)

  if bool(class_points_tuple[0]) != bool(class_points_tuple[1]):
    class_points_output.append(1)
  else:
    class_points_output.append(0)
  tuple_xy = (aux1, aux2)
  print(f'Posições no eixo (x, y) dos dados de entrada: {tuple_xy}')
  print(f'Classes das classes de entrada: {class_points_tuple}.')
  print(f'Após cálculo da porta XOR das classes de entrada: {class_points_output[count]}')
  axle_points_vector.append(tuple_xy)
  print('\n')

new_point_x = random.uniform(1, 10)
new_point_x = round(new_point_x,2)
new_point_y = random.uniform(1, 10)
new_point_y = round(new_point_y,2)
new_point = (new_point_x, new_point_y)


euclidian_distances_vector = []
for count in range(10):
  euclidiano_ponto_a = np.array(axle_points_vector[count])
  euclidiano_ponto_b = np.array(new_point)
  
  euclidian_distance = np.linalg.norm(euclidiano_ponto_a - euclidiano_ponto_b)
  euclidian_distances_vector.append(round(euclidian_distance,4))
  print(f'As distâncias euclidianas entre o ponto A{count}={axle_points_vector[count]} e o ponto B={new_point} é igual a: {euclidian_distances_vector[count]}.')

print()
k_nearest_points = []
k_nearest_points = deepcopy(euclidian_distances_vector)
k_nearest_points = sorted(k_nearest_points)

for count in range(k):
  nearest = k_nearest_points[count]
  index = euclidian_distances_vector.index(nearest)
  if class_points_output[index] == 0:
    print(f'O {count+1}o vizinho mais proximo do novo ponto pertence a classe 0.')
    zeros+=1
  else:
    print(f'O {count+1}o vizinho mais proximo do novo ponto pertence a classe 1.')
    ones+=1
if zeros>ones:
  print(f'O novo ponto {new_point} pertence a classe 0.')
  class_points_output.append(0)
else:
  print(f'O novo ponto {new_point} pertence a classe 1.')
  class_points_output.append(1)
```
